```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("netflix_titles.csv", encoding='utf-8')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
df.head()
```

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.head()
```

```python
df['country'] = df['country'].str.title()
df['listed_in'] = df['listed_in'].str.title()
df['description'] = df['description'].str.title()
df.head()
```

```python
# Step 1: Replace 'Unknown' with NaN to parse dates properly
df['date_added'] = df['date_added'].replace('Unknown', pd.NA)

# Step 2: Convert to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Step 3: Fill missing dates again as 'Unknown'
df['date_added'] = df['date_added'].fillna('Unknown')

# Step 4 : Format as dd-mm-yyyy string if not Unknown
df['date_added'] = df['date_added'].apply(
    lambda x: x.strftime('%d-%m-%Y') if x != 'Unknown' else x
)

df.head()
```

```python
def parse_duration(x):
    if 'Seasons' in x:
        return pd.Series([int(x.split()[0]), 'Seasons'])
    elif 'Season' in x:
        return pd.Series([int(x.split()[0]), 'Season'])
    elif 'min' in x:
        return pd.Series([int(x.split()[0]), 'min'])
    else:
        return pd.Series([None, None])

df[['duration_num', 'duration_type']] = df['duration'].apply(parse_duration)
```
